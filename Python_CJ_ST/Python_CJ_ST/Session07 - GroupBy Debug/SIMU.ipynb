```python
import math
import numpy as np
import pandas as pd
```

### csv에서 음수는 1로 치환

```python
selloutData = pd.read_csv("../dataset/kopo_channel_seasonality_new.csv")
selloutData.head(5)
```

```python
selloutData["QTY_NEW"] = np.where(selloutData["QTY"]<0,1,selloutData["QTY"])
selloutData.head()
```

## YEARWEEK을 STR로 변경하고 인덱스4~6의 53외 값만 취하라

```python
selloutData = selloutData.loc[selloutData["YEARWEEK"].astype(str).str[4:6]!="53"]
selloutData.head()
```

```python
len(selloutData)
```

## YEAR하고 WEEK 분리

```python
selloutData["YEAR"] = selloutData["YEARWEEK"].astype(str).str[0:4]
selloutData["WEEK"] = selloutData["YEARWEEK"].astype(str).str[4:6]
selloutData.head()
```

## DATA 정렬하기

```python
selloutData = selloutData.sort_values(["REGIONID","PRODUCT","YEARWEEK"], ascending= [True,False,False])
selloutData.head()
```

## INDEX RESET

```python
selloutData = selloutData.reset_index(drop=False)
selloutData.head()
```

## DATA 분석

## 1. 평균

```python
groupData = selloutData.groupby(["REGIONID","PRODUCT"], as_index=False).mean()
groupData.head()
```
