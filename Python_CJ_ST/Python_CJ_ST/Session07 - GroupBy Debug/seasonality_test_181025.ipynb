### 그룹바이 함수 구현

```python
def rolling1(eachGroup):
    # 인덱스 초기화
    logic1 = eachGroup.reset_index(drop=True)

    # 이동평균 컬럼 생성
    logic1["MA"] = logic1.QTY.rolling(window=5, min_periods=1 , center=True).mean()

    # 최종
    return logic1
```

```python
groupResult = sortedData.groupby(["REGIONID","PRODUCT"]).apply(rolling1)
```

```python
finalResult = groupResult.reset_index(drop=True)
```

```python
finalResult["RATIO"] = finalResult.QTY/finalResult.MA
```
