### 4. 이상 데이터 정제하기

```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine 
 
# CSV 파일을 읽어 Data Frame 변수에 저장하기
customerData = pd.read_csv("../dataset/customerdata.csv")

customerData.columns = ["CUSTID","AVGPRICE","EMI","DEVICECOUNT","PRODUCTAGE","CUSTTYPE"]
#customerData.columns = [x.upper() for x in customerData.columns]

# 데이터 VIEW
customerData.head()


customerData["PRODUCTAGE"]= np.\
where(customerData["PRODUCTAGE"] < 1, 1,
customerData["PRODUCTAGE"])

customerData.head()
```

### [추가] 데이터 컬럼간 연산하기

```python
customerData["NEW_COLUMN"] = customerData["EMI"].astype(float) * customerData["DEVICECOUNT"]
customerData.head()
```

### 실습문제

```python
def refining(df):
    if df['PRODUCTAGE'] < 1:
        return 1
    elif df['PRODUCTAGE'] < 2:
        return 2
    elif df['PRODUCTAGE'] < 3:
        return 3
    else:
        return 5
```

```python
customerData["PRODUCTAGE_NEW2"]=customerData.apply(refining, axis=1)
```

```python
customerData.head()
```

```python
# numpy 라이브러리 
import numpy as np

#컬럼 추가 생성 및 조건에 따른 컬럼명 변경
customerData["PRODUCTAGE_NEW"] = np.\
where(customerData["PRODUCTAGE"] < 3, 1, customerData["PRODUCTAGE"])
customerData.head()
```

# B. 데이터 합치기

```python
import pandas as pd
from sqlalchemy import create_engine 
 
# CSV 파일을 읽어 Data Frame 변수에 저장하기
selectExample = pd.read_csv("../dataset/kopo_channel_seasonality_new.csv")
print(selectExample.head())

# CSV 파일을 읽어 Data Frame 변수에 저장하기
regionMasterData = pd.read_csv("../dataset/kopo_region_mst.csv", encoding = "utf-8-sig") 

print(regionMasterData.head())
```

```python
mergeResult = pd.merge(selectExample, regionMasterData,\
        left_on="REGIONID", right_on = "REGIONID", how="inner")\
[["REGIONID","REGIONNAME","PRODUCT","YEARWEEK","QTY"]]
mergeResult.head()
```

```python
len(mergeResult)
```

```python
import numpy as np

# A01 데이터 정제
a01data = selectExample[(selectExample['REGIONID']=='A01')]
print(len(a01data))

# A71 데이터 정제
a71data = selectExample[(selectExample['REGIONID']=='A02')]
a71data['QTY'] = np.where(a71data['QTY'] < 500,500,a71data['QTY'])
print(len(a71data))

# 데이터 결과 값 MERGE
mergedata = pd.concat([a01data, a71data], axis=0)
print(len(mergedata))
```
